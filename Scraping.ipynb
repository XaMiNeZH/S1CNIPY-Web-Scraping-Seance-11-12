# Section N°1:
1. Chercher le contenu du Site Web
2. Extraire le contenu exploitable avec BeautifulSoup
3. Extraire les données souhaitées

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import os
```

```python
URL = 'https://books.toscrape.com/'
response = requests.get(URL)
#dir(response)
response
```

```python
page_contents = response.text
#page_contents
```

```python
contenu=BeautifulSoup(page_contents,"html.parser")
#contenu
```

```python
def get_book_titles(doc):
    Book_title_tags = doc.find_all('h3')
    Book_titles = []
    for tags in Book_title_tags:
        Book_titles.append(tags.text)
    return Book_titles
get_book_titles(contenu)
```

```python
def get_book_price(doc):  
    Book_price_tags = doc.find_all('p', class_ = 'price_color')  
    Book_price = []  
    for tags in Book_price_tags:  
        Book_price.append(tags.text.replace('Â£','€'))  
    return Book_price  
  
get_book_price(contenu)
```

```python
def get_stock_availability(doc):
    Book_stock_tags = doc.find_all('p', class_='instock availability')
    Book_stock = []
    for tags in Book_stock_tags:
        Book_stock.append(tags.text.strip())
    return Book_stock
get_stock_availability(contenu)
```

```python
def get_book_url(doc):
    Book_url = []
    Book_title_tags = doc.find_all('h3')
    for article in Book_title_tags:
        for link in article.find_all('a', href= True):
            url = link['href']
            links = 'https://books.toscrape.com/' + url
            if links not in Book_url:
                Book_url.append(links)
    return Book_url
get_book_url(contenu)
```

# Section N°2:
1. Chercher le contenu de plusieurs pages du Site Web
2. Regrouper les listes des données dans dictionnaire

```python
URL = 'https://books.toscrape.com/'
def Cherche_contenu(Url,data):
    response = requests.get(Url)
    contenu=BeautifulSoup(response.text,"html.parser")
    match data:
        case 'title': return get_book_titles(contenu)
        case 'price': return get_book_price(contenu)
        case 'dispo': return get_stock_availability(contenu)
        case 'link': return get_book_url(contenu)
        case _: return None
```

```python
Title=[]
for k in range(1,6):
    L=Cherche_contenu(f'https://books.toscrape.com/catalogue/page-{k}.html',"title")
    Title.extend(L)
print(len(Title))
#print(len(set(Title)))
```

```python
Price=[]
for k in range(1,6):
    L=Cherche_contenu(f'https://books.toscrape.com/catalogue/page-{k}.html',"price")
    Price.extend(L)
print(len(Price))
#print(len(set(Price)))
```

```python
Dispo=[]
for k in range(1,6):
    L=Cherche_contenu(f'https://books.toscrape.com/catalogue/page-{k}.html',"dispo")
    Dispo.extend(L)
print(len(Dispo))
#print(len(set(Dispo)))
```

```python
Link=[]
for k in range(1,6):
    L=Cherche_contenu(f'https://books.toscrape.com/catalogue/page-{k}.html',"link")
    Link.extend(L)
print(len(Link))
#print(len(set(Link)))
```

```python
Data={
    "Title":Title,
    "Price":Price,
    "avalaible":Dispo,
    "Link":Link}
Data
```
